# Problems generation

```python
from src import read_json, write_json

questions = read_json('./data/questions.json')
locations = read_json('./data/locations.json')
```

```python
TEMPLATE = "{location}. {question}"

problems = []

for question in questions:
    for location in locations:
        problem = TEMPLATE.format(location=location, question=question)
        problems.append(problem)

len(problems)
```

```python
write_json(problems, './data/problems.json')
```
