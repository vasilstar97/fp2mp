# Problems generation

```python
QUESTIONS_FILE = './data/questions.json'
LOCATIONS_FILE = './data/locations.json'
PROBLEMS_FILE = './data/problems.json'
```

```python
from src import read_json, write_json, generate_id

questions = read_json(QUESTIONS_FILE)
locations = read_json(LOCATIONS_FILE)
```

```python
TEMPLATE = "{location}. {question}"

problems = {}

for question in questions:
    for location in locations:
        problem_id = generate_id(location, question)
        problem_content = TEMPLATE.format(location=location, question=question)
        problems[problem_id] = {
            'location': location,
            'question': question,
            'content': problem_content
        }

len(problems)
```

```python
write_json(problems, PROBLEMS_FILE)
```
